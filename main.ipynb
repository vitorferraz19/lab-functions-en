## 2. Now write a function that returns the largest element on a list

```python
def greatest(numbers):
    largest = numbers[0]              # initializing via the assumption that the biggest number is at the 0th index on the list

    for number in numbers:            # variable number will keep track of the largest number so far in the sequence
        if number > largest:
            largest = number          #at this point, it means we have found the largest number, based on the specified condition

    return largest
```

```python
# This will test your function 
test_greatest(greatest)
```

## 3. Write a function that sums all the elements on a list

```python
def sum_all(lst):
    total = 0                     # initializing via a 0 total
    
    for number in lst:            # variable number will keep track of the total so far in the sequence
        total += number          # calculating the sum total

    return total
```

## 4. Write another function that multiplies all the elements on a list

```python
def mult_all(lst):
    total = 1                     # initializing via a 1 total for the multiplication case (not 0, otherwise we're multiplying by 0 is itself 0)
    
    for number in lst:            # variable number will keep track of the mult. total so far in the sequence
        total *= number          # calculating the multiplication total

    return total
```

## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    print(n, '!')
    if n <= 0:                        # a corner case where we reach 0
        return 1                      # given that 0! is 1
    return n * factorial(n - 1)       # this return runs outside of the if condition above; recursive use of the function from n, n-1, etc
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers 
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop, 
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n. 
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

```python
# This will test your function 
test_factorial(factorial)
```
