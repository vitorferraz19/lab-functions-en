## 2. Now write a function that returns the largest element on a list

```python
def greatest(numbers):
    largest = numbers[0]              # initializing via the assumption that the biggest number is at the 0th index on the list

    for number in numbers:            # variable number will keep track of the largest number so far in the sequence
        if number > largest:
            largest = number          #at this point, it means we have found the largest number, based on the specified condition, so we force return of the largest number

    return largest
```

## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    print(n, '!')
    if n <= 0:                        # a corner case where we reach 0
        return 1                      # given that 0! is 1
    return n * factorial(n - 1)       # this return runs outside of the if condition above; recursive use of the function from n, n-1, etc
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers 
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop, 
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n. 
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

## 7. Write a function that takes a list and returns a list of the unique values.

`NOTE: You cannot use set. 🤔`

```python
def unique(lst_un):
    
    list_uniquenumb = []          # initialising via an empty list in which values will be appended                   
    
    for element in lst_un:
        if element not in list_uniquenumb:      # meaning, for every element in the input that has not been appended yet          
            list_uniquenumb.append(element)           # keep appending

    return list_uniquenumb
```

## 8. Write a function that returns the mode of a list, i.e.: the element that appears the most times.
`NOTE: You should not use count... 🧐`

```python
def mode_counter(arr):
    
    element_counts = {}          # initialising via an empty dictionary (to map keys and corresponding values) in which values will be appended                   
    
    for element in arr:               # counting how many times an element occurs
        if element in element_counts:      # meaning, for every element in the array         
            element_counts[element] += 1           # keep adding in case this element is repeated

        else:
            element_counts[element] = 1

        maximum_count = 0
        mode = None

    for key in element_counts:                       # detecting the highest occurence of a particular key in the dictionary (i.e., the mode)
        if element_counts[key] > maximum_count:        # here we put an if condition to detect the maximum count of a given key
            maximum_count = element_counts[key]
            mode = key
    
    return mode
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. 😉`

```python
def st_dev(list_sd):

#given the definition of standard deviation, we can first calculate the mean, then subtract mean from each value, calculate the variance and then extract its square root, which is the standard deviation itself

    if len(list_sd) == 0:
        return None

    total = 0     #initializing mean from 0

    for number in list_sd:
        total += number

    mean = total / (len(list_sd)-1)      # calculating mean

    sum_differences_squared = 0    #initializing difference sums from 0

    for number in list_sd:
        diff = number - mean
        squared = diff ** 2
        sum_differences_squared += squared

    variance = sum_differences_squared / (len(list_sd)-1)      # calculating the variance

    return variance ** 0.5     # i.e., the square root of the variance, without using a math library function (sqrt)

    
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! 🤪`

```python
def sort_alpha(string):

    words = []                      #initialising via an empty list of words
    present_word = ""               #initialising via an empty word

    for character in string:
        if character == ",":       #checking comma separation
            words.append(present_word)
            present_word = ""                  # ensuring parenthesis in the word again
        else:
            present_word += character       #keep adding character otherwise

    words.append(present_word)

    words = sorted(words)                  #sorting the words (allowed)

    resulting_string = ""                #initialising via an empty word

    for i in range(len(words)):
        resulting_string += words[i]
        if i != len(words) - 1:
            resulting_string += ','

    return resulting_string
```

```python
# This will test your function 
test_alpha(sort_alpha)
```
